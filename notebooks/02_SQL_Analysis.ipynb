## count with group by

```python
q="select region,count(*) total_orders from sales group by region"
pd.read_sql(q,conn)
```

```python
q = """
SELECT Region,
       COUNT(*) AS Total_Orders
FROM sales
GROUP BY Region
HAVING COUNT(*) > 1000
"""

pd.read_sql(q, conn)
```

## Insight:
East, West, and Central have more than 1000 orders.
South has fewer than 1000 orders and may require business attention.

```python
q = """
SELECT Category,
       AVG(Sales) AS Avg_Sales
FROM sales
GROUP BY Category
HAVING AVG(Sales) > 30
"""

pd.read_sql(q, conn)
```

## Insight:
Technology has the highest average sales per transaction, indicating stronger revenue generation per order.

Furniture also maintains an average sales value above 30, making it an important revenue contributor.

```python
q = """
SELECT Region,
       SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Region
ORDER BY Total_Sales DESC
LIMIT 3
"""

pd.read_sql(q, conn)
```

## Insight:
West region generated the highest total sales, followed by East and Central.

These regions contribute significantly to overall business revenue and should be prioritized for growth strategies.

# Key Business Findings

1. West region generated the highest sales and order volume.
2. Technology category achieved the highest average profit.
3. South region had the lowest order count.
4. Furniture and Technology categories recorded average sales above 30.
5. Office Supplies and Technology maintained average profits greater than 5.

## CTE-Common Table Expression

```python
q="with regionsales as (select region,sum(sales) as total_sales from sales group by region) select *from regionsales"
pd.read_sql(q,conn)
```

## Top Region Using CTE

```python
q="with regionsales as (select region,sum(sales) as total_sales from sales group by region) select *from regionsales order by total_sales desc limit 1"
pd.read_sql(q,conn)
```

## Window Functions

   ## ROW_NUMBER()

```python
q = """
SELECT Region,
       Sales,
       ROW_NUMBER() OVER
       (
           ORDER BY Sales DESC
       ) AS Row_Num
FROM sales
"""
pd.read_sql(q, conn)
```

  ## RANK()

```python
q = """
SELECT Region,
       Sales,
       RANK() OVER
       (
           ORDER BY Sales DESC
       ) AS Sales_Rank
FROM sales
"""
pd.read_sql(q, conn)
```

## LAG()

```python
q="select region,sales,lag(sales) over(order by sales desc) as previous_sale from sales"
pd.read_sql(q, conn)
```

## LEAD()

```python
q="select region,sales,lead(sales) over(order by sales desc) as Next_sale from sales"
pd.read_sql(q, conn)
```

## Create a view - HighSales

```python
conn.execute("DROP VIEW IF EXISTS HighSales")

q = """
CREATE VIEW HighSales AS
SELECT *
FROM sales
WHERE Sales > 50 
"""
conn.execute(q)
print("View created successfully!")
```
