```python
import sqlite3
import pandas as pd
conn=sqlite3.connect("cleaned_superstore.db")
print("Database connected sucessfully")
```

## Step 2: Verify Available Tables

Before querying data, it is important to check which tables exist in the database.

```python
q = """
SELECT name
FROM sqlite_master
WHERE type='table'
"""

pd.read_sql(q, conn)
```

## Step 3: Business Questions & Insights


## Business Question 1: Top 5 Products by Total Sales

### Objective
Identify the products generating the highest sales revenue.

```python
q="""
SELECT
    [Product Name],
    SUM(Sales) AS Total_Sales
FROM sales
GROUP BY [Product Name]
ORDER BY Total_Sales DESC
LIMIT 5
"""

pd.read_sql(q,conn)
```

## Business Question 2: Sales by Region

### Objective
Determine which region contributes the highest sales.

```python
q="""
SELECT
    Region,
    SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Region
ORDER BY Total_Sales DESC
"""

pd.read_sql(q,conn)
```

## Business Question 3: Category Performance

### Objective
Compare sales across product categories.

```python
q="""
SELECT
    Category,
    SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Category
ORDER BY Total_Sales DESC
"""

pd.read_sql(q,conn)
```

## Business Question 4: Most Profitable Category

### Objective
Identify categories generating the highest profit.

```python
q="""
SELECT
    Category,
    SUM(Profit) AS Total_Profit
FROM sales
GROUP BY Category
ORDER BY Total_Profit DESC
"""

pd.read_sql(q,conn)
```

## Business Question 5: Average Order Value by Region

### Objective
Measure average sales value per order in each region.

```python
q="""
SELECT
    Region,
    AVG(Sales) AS Avg_Order_Value
FROM sales
GROUP BY Region
ORDER BY Avg_Order_Value DESC
"""

pd.read_sql(q,conn)
```

## Business Question 6: Top 10 Customers by Sales

### Objective
Identify the customers generating the highest revenue.

```python
q="""
SELECT
    [Customer Name],
    SUM(Sales) AS Total_Sales
FROM sales
GROUP BY [Customer Name]
ORDER BY Total_Sales DESC
LIMIT 10
"""
pd.read_sql(q,conn)
```

## Business Question 7: Sales by Customer Segment

### Objective
Compare sales across customer segments.

```python
q="""
SELECT
    Segment,
    SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Segment
ORDER BY Total_Sales DESC
"""
pd.read_sql(q,conn)
```

## Business Question 8: Top Cities by Sales

### Objective
Identify cities generating the highest revenue.

```python
q="""
SELECT
    City,
    SUM(Sales) AS Total_Sales
FROM sales
GROUP BY City
ORDER BY Total_Sales DESC
LIMIT 10
"""
pd.read_sql(q,conn)
```

## Business Question 9: Average Profit by Discount Level

### Objective
Analyze how discounts affect profitability.

```python
q="""
SELECT
    Discount,
    AVG(Profit) AS Avg_Profit
FROM sales
GROUP BY Discount
ORDER BY Discount
"""
pd.read_sql(q,conn)
```
